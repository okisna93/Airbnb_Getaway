```python
import folium
import pandas as pd
```

```python
coordinates=[42.251888, -73.948056]
site_map=folium.Map(location=coordinates,zoom_start=6)
site_map
```

```python
cabins_data=pd.read_csv("cabins_location.txt",header=None)
cabins_data=cabins_data.reset_index()
cabins_data.rename(columns={'index':'CabinNumber',0:'Lat',1:'Long',2:'CabinName'},inplace=True)
cabins_data.head(5)
```

```python
cabins_data.columns
```

```python
df_reorder=cabins_data[['CabinNumber','CabinName','Lat', 'Long']]
cabins_data=df_reorder
cabins_data.head()
```

```python
labels=[]
for name, number in zip(cabins_data['CabinName'],cabins_data['CabinNumber']):
    labels.append(name+" - "+str(number))
labels
```

## Below can be used to add different map types

 folium.TileLayer('Stamen Terrain').add_to(site_map)
 folium.TileLayer('Stamen Toner').add_to(site_map)
 folium.TileLayer('Stamen Water Color').add_to(site_map)
 folium.TileLayer('cartodbpositron').add_to(site_map)
 folium.TileLayer('cartodbdark_matter').add_to(site_map)
 folium.LayerControl().add_to(site_map)

```python
incidents=folium.map.FeatureGroup()

for lat,long,label in zip(cabins_data['Lat'],cabins_data['Long'],labels):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat,long],
            radius=10,
            color='red',
            fill=True,
            popup=label,
            fill_color='green',
            fill_opacity=0.6
        )
    ).add_to(site_map)
    
site_map
```
